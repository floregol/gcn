```python
import numpy as np
import pickle as pk
from utils import *
import matplotlib.pyplot as plt
eigen_file_name = "shape_2708.p"
  
with open(eigen_file_name, 'rb') as f:
    svd_decomposition = pk.load(f, encoding='latin1')
    
eigenval = svd_decomposition['s']
eigenvectors = svd_decomposition['vh']
adj, features,labels, y_train, y_val, y_test, initial_train_mask, val_mask, test_mask = load_data(
        "cora")
```

```python
def eigen(x):
    return eigenvectors * x
y= np.abs(eigen(labels))
```

```python
powers = []
for j in range(7):
    power = []
    cdf =0 
    for i in range(y.shape[0]):
        cdf += np.square(y[i,j])
        power.append(cdf)
    power = power/cdf
    powers.append(power)
powers = np.sum(powers, axis=0)
powers = powers/7
```

```python
print(powers.shape)
plt.plot(powers)

plt.axis([0,2710,0,1.01])
plt.savefig("power.png",dpi=300)
```
