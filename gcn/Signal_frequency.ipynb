```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pk
from utils import load_data
```

```python
adj, features, y_train, y_val, y_test, initial_train_mask, val_mask, test_mask = load_data(
        'cora')
```

```python

y = y_train+y_val+y_test
print(y[:,0:1].shape)
```

```python

eigen_file_name = 'shape_2708.p'
with open(eigen_file_name, 'rb') as f:
    svd_decomposition = pk.load(f, encoding='latin1')
eigenval = svd_decomposition['s']
eigenvectors = svd_decomposition['vh']
print(eigenval)
```

```python
z = np.ones((2708, 1))
print(z)
```

```python
def getFourierTransform(x):
    return eigenvectors * x
```

```python
z = getFourierTransform(y)
```

```python
print(z)
```
